# Проверка ЦПД (распределение Пуассона)

Посмотрим, как ведет себя конечная выборка и теоретическая зависимость:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

s = np.random.normal(size=10000)

plt.hist(s, 30, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')
plt.show()

# Рисуем график самого распределения
x = np.arange(-10, 10, 0.01)
plt.plot(x, norm.pdf(x, 0, 1))
```

Сгенерируем выборку чисел по распределению Пуассона и нанесем ее на график:

```python
import numpy as np
import matplotlib.pyplot as plt

vib = np.array([])

# Выборка размером 5
for i in range(1000):
    n = 5
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')
plt.show()

# Выборка размером 10
for i in range(1000):
    n = 10
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')
plt.show()

# Выборка размером 50
for i in range(1000):
    n = 50
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')
plt.show()

# Выборка размером 1000
for i in range(1000):
    n = 1000
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')
plt.show()
```

Дисперсия распределения Пуассона равна его матожиданию и равна $\lambda$. В нашем случае $\lambda = 5$. Отсюда, для нормального распределения, к которому мы стремимся, имеем: $\mathbb{M} = \boxed{5 = \mu}; \mathbb{D} = 5 = \sigma^2 \Rightarrow \boxed{\sigma = \sqrt{5}}$. На основании этого рисуем Гауссово распределение поверх наших гистограмм.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

vib = np.array([])

# Выборка размером 5
for i in range(1000):
    n = 5
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')

# Рисуем график самого распределения
x = np.arange(-10, 10, 0.01)
plt.plot(x, norm.pdf(x, 5, 0.95))

plt.show()

# Выборка размером 10
for i in range(1000):
    n = 10
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')

# Рисуем график самого распределения
x = np.arange(-10, 10, 0.01)
plt.plot(x, norm.pdf(x, 5, 0.85))

plt.show()

# Выборка размером 50
for i in range(1000):
    n = 50
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')

# Рисуем график самого распределения
x = np.arange(-10, 10, 0.01)
plt.plot(x, norm.pdf(x, 5, 0.6))

plt.show()

# Выборка размером 1000
for i in range(1000):
    n = 1000
    s = np.random.poisson(5, n)
    vib = np.append(vib, np.mean(s))
    
plt.hist(vib, 14, normed=True)
plt.ylabel('p(k)')
plt.xlabel('k')

# Рисуем график самого распределения
x = np.arange(-10, 10, 0.01)
plt.plot(x, norm.pdf(x, 5, 0.4))

plt.show()
```

Видно, что гистограммы приближаются к нормальному распределению, как и должно быть согласно ЦПД. Более того, с ростом n точность растет
